## Liar dataset

```python
Liar_preprocessed = pd.read_table('https://raw.githubusercontent.com/tfs4/liar_dataset/master/train.tsv')
Liar_preprocessed.columns =['json_id', 'claim', 'object', 'topic', 'speaker', 'job_title (string)','state_info (string)','party_affiliation (string)','barely_true_counts (float32)','false_counts (float32)','half_true_counts (float32)','	mostly_true_counts (float32)','	pants_on_fire_counts (float32)','context (string)']
Liar_computed = Liar_preprocessed[['json_id', 'claim','object']].copy()

```

```python
print(Liar_computed)
```

```python
def condition(x):
    if x =='false':
        return 1
    elif x== 'half-true':
        return 2
    elif x== 'mostly-true':
        return 2
    elif x=='true':
        return 0
    elif x =='barely-true':
        return 2
    elif x == 'pants-fire':
        return 1


Liar_computed['binary label'] = Liar_computed['claim'].apply(condition)
print(len(Liar_computed))
```

```python
import matplotlib.pyplot as plt

claims = ['false','true', 'half-true','mostly-true','barely-true', 'pants-fire']

def count_claims(df, claim):
    count_claim = len(df[df['claim']== claim])
    return count_claim

claim_list= []
for x in claims:
    count_claim = count_claims(Liar_computed,x)
    claim_list.append(count_claim)
    print(f"Number of {x} statements: {count_claim}")


binary_labels = [0,1,2]
def count_label(df, labels):
    count_label = len(df[df['binary label']== labels])
    return count_label

claim_labels = []
for x in binary_labels:
    count_labels = count_label(Liar_computed,x)
    claim_labels.append(count_labels)
    print(f"Number of {x} statements: {count_labels}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle('Distribution of Claims and Labels')

ax1.pie(claim_list, labels=claims)
ax1.set_title('Distribution of Claims')

ax2.pie(claim_labels, labels=['True claims', 'False claims', 'In between claims'])
ax2.set_title('Distribution of Labels')

plt.show()
```

## Computing the different distributions

```python
from faKy import values_by_label, compute_statistics
labels = [0, 1, 2]
df_label = 'binary label'
```

### Complexity

```python
from faKy import process_text_complexity


Liar_computed['compressed_size'] = Liar_computed['object'].apply(process_text_complexity)
```

```python
label_com = values_by_label(Liar_computed, 'compressed_size',labels,df_label)
stat, p = kruskal(*label_com) # unpack the elements 

print('Kruskal results:')
print(f'F-statistic: {stat:.3f}')
print(f'p-value: {p:.3e}')
```

```python
dunn_results = sp.posthoc_dunn(label_com, p_adjust='bonferroni')

print('Dunn results for the information complexity')
dunn_table(dunn_results)
```

### Readability

### VADER Sentiment

```python
from faKy import process_text_vader
Liar_computed[['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound']] = Liar_computed['object'].apply(
                                                                        lambda x: pd.Series(process_text_vader(x)))

```

```python
Liar_computed.head()
```

```python
vader_labels = ['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound']

computed_labels = []
for vader in vader_labels:
    label_vader = values_by_label(Liar_computed, vader,labels,df_label)
    stat, p = kruskal(*label_vader) # unpack the elements
    computed_labels.append([stat, p])

for i, (stat, p), vader in zip(range(len(computed_labels)), computed_labels, vader_labels):
    print(f'{vader} : fstat {stat}; p {p}')
```

```python
for vader in vader_labels:
    label_vader = values_by_label(Liar_computed, vader,labels,df_label)
    dunn_results = sp.posthoc_dunn(label_vader, p_adjust='bonferroni')
    a = dunn_table(dunn_results)
    print(f'Dunn results: {vader}')
    print(a)
```

### NER

```python
from faKy import count_named_entities,count_ner_labels, create_input_vector_NER, ner_labels

Liar_computed['tot_ner_count'] = Liar_computed['object'].apply(count_named_entities)
Liar_computed['ner_counts'] = Liar_computed['object'].apply(count_ner_labels)
Liar_computed['input_vector_ner'] = Liar_computed['ner_counts'].apply(create_input_vector_NER)

for tag in ner_labels:
    col_name = f'NER_{tag}'
    Liar_computed[col_name] = Liar_computed['input_vector_ner'].apply(
                                    lambda x: x[ner_labels.index(tag)] if tag in ner_labels else 0)
Liar_computed.columns
```

```python
Liar_computed.head()
```

```python
label_NER = values_by_label(Liar_computed, 'tot_ner_count', labels, df_label)
stat, p = kruskal(*label_NER) # unpack the elements 

print('Kruskal results:')
print(f'F-statistic: {stat:.3f}')
print(f'p-value: {p:.3e}')
```

```python
dunn_results = sp.posthoc_dunn(label_NER, p_adjust='bonferroni')


print('Dunn results for the NER')
dunn_table(dunn_results)
```

```python
entity_counts = Liar_computed.groupby('binary label')['tot_ner_count'].sum()

binary_labels = [0,1,2]
def count_label(df, labels):
    count_label = len(df[df['binary label']== labels])
    return count_label

N = len(Liar_computed)
true = count_label(Liar_computed,0)
false = count_label(Liar_computed,1)
between = count_label(Liar_computed,2)

total_entities_true = entity_counts[0]
total_entities_false = entity_counts[1]
total_entities_inbetween = entity_counts[2]



true_ratio = np.rint((true/total_entities_true)*100)
false_ratio =  np.rint((false/ total_entities_false)*100)
between_ratio =  np.rint((between/ total_entities_inbetween)*100)

print(f'total_entities_true: {true_ratio}%')
print(f'total_entities_false: {false_ratio}%')
print(f'total_entities_inbetween: {between_ratio}%')
```

```python
ner_labels = [ 'NER_CARDINAL',
       'NER_DATE', 'NER_EVENT', 'NER_FAC', 'NER_GPE', 'NER_LANGUAGE',
       'NER_LAW', 'NER_LOC', 'NER_MONEY', 'NER_NORP', 'NER_ORDINAL', 'NER_ORG',
       'NER_PERCENT', 'NER_PERSON', 'NER_PRODUCT', 'NER_QUANTITY', 'NER_TIME',
       'NER_WORK_OF_ART']
computed_ner_labels = []
for ner in ner_labels:
    label_ner = values_by_label(Liar_computed, ner, labels,df_label)
    stat, p = kruskal(*label_ner) # unpack the elements
    computed_ner_labels.append([stat, p])

for i, (stat, p), pos in zip(range(len(computed_ner_labels)), computed_ner_labels, ner_labels):
       print(f'{pos} : fstat {stat}; p {p}')
```

```python
for ner in ner_labels:
    label_ner = values_by_label(Liar_computed, ner,labels, df_label)
    dunn_results = sp.posthoc_dunn(label_ner, p_adjust='bonferroni')
    a = dunn_table(dunn_results)
    print(f'Dunn results: {ner}')
    print(a)
```

points of interest:
